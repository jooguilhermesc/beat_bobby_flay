Inicialmente eu queria fazer duas análises, qual o convidado que mais venceu Bobby e qual a escola culinária que o Bobby tinha mais dificuldade, infelizmente essa segunda análise seria mais complexa, porque eu não consegui encontrar um dataset que eu pudesse utilizar para enriquecer o dataset com as infos do programa, então me fixei em fazer uma análise de quais duplas de convidados mais venceu Bobby e qual convidado, individualmente, mais venceu Bobby.

Como análise é feito utilizando Pandas, a primeira ação foi importar a biblioteca

```python
import pandas as pd
```

Depois de importar a biblioteca agora é hora de começar a trabalhar o dataset que vai me ajudar nessa análise. Utilizando a função read_html() passei o link da página da wikipedia. Essa função ajuda a extrair dados de tabelas presentes em uma página e retorna um array com todas as tabelas.

```python
df_battles = pd.read_html('https://en.wikipedia.org/wiki/Beat_Bobby_Flay')
```

Com os dados armazenados na variável df_battles agora é hora de ver o que está no array das tabelas e tentar encontrar as tabelas que nos interessam.

```python
df_battles[0:3]
```

Percebemos que essa primeira extração retorna um array de tabelas diversas, não só as tabelas com infos dos episódios, agora preciso analisar a partir de qual posição desse array começamos a ter as tabelas com essas infos. Analisando a primeira posição, percemos que retorna dados gerais sobre o programa.

```python
df_0 = df_battles[0]
df_0.head(5)
```

Mas a partir da segunda posição do array já notamos algumas informações sobre as temporadas

```python
df_1 = df_battles[1]
df_1.head(5)
```

Entretanto, as tabelas estão voltando com múltiplos leveis de colunas e pra poder trabalhar com elas precisamos remover esse primeiro level

```python
df_1 = df_battles[1]
df_1.columns = df_1.columns.droplevel(0)
columns = ['No.','Title', 'Original airdate', 'Guest(s)', 'Ingredient(s)','Contestants', 'Judges', 'Dish', 'Winner']
df_test = df_1[columns]
df_test.head(5)
```

Agora o problema dos múltiplos leveis das colunas foi resolvido, preciso ajustar a coluna "Guest(s)" para retornar os nomes dos convidados em duas colunas separadas "Guest A" e "Guest B", que mais tarde será utilizada para saber qual convidado individualmente mais venceu Bobby. Não vou excluir a colunas "Guest(s)" porque também quero saber qual dupla mais venceu Bobby Flay.

```python
array_guests = df_test["Guest(s)"].str.split(",",n=1).str
df_test["Guest A"] = array_guests[0]
df_test["Guest B"] = array_guests[1]
df_guests = df_test
df_guests.head(5)
```

Assim como para os convidados farei o tratamento de separar os jurados também.

```python
array_judges = df_guests["Judges"].str.split(",",n=2).str
df_guests["Judge A"] = array_judges[0]
df_guests["Judge B"] = array_judges[1]
df_guests["Judge C"] = array_judges[2]
df_judges = df_guests
df_judges.head(5)
```

Ainda farei uma modificação na coluna "Constestants" apenas para apresentar a batalha de uma maneira esteticamente mais agradável.

```python
df_judges["Contestants"] = df_judges["Contestants"].apply(lambda x: x.replace(","," X "))
df_judges.head(5)
```

E acrescento a informação da temporada a qual aquela batalha pertence, nesse caso em específico, estou analisando os dados da primeira temporada.

```python
df_judges["Season"] = 1
```

Depois de acrescentadas essas informações, meu dataframe ficará dessa forma

```python
df_judges.head(5)
```

Finalizado os tratamentos que quero fazer, posso exportar essa informação para um arquivo CSV, que posteriormente será enriquecido com os dados das outras temporadas.

```python
df_judges.to_csv('csv/df_bbf.csv',index=False, encoding="utf-8")
```

Agora para enriquecer o nosso dataset vou iterar sobre o df_battles e extrair as informações das demais temporadas e armazenando no csv já criado, pra isso farei uma pequena modificação na função to_csv() acrescentando os parâmetros mode="a" e header=False, isso fará com que no momento de inserir no nosso arquivo csv os dados sejam anexados e não tenhamos linhas de cabeçalho como dados em nosso dataset.

```python
for i in range(2,len(df_battles)-1):
    try:
        df_test = df_battles[i]
        df_test.columns = df_test.columns.droplevel(0)
        array_guests = df_test["Guest(s)"].str.split(",",n=1).str
        df_test["Guest A"] = array_guests[0]
        df_test["Guest B"] = array_guests[1]
        df_guests = df_test
        array_judges = df_guests["Judges"].str.split(",",n=2).str
        df_guests["Judge A"] = array_judges[0]
        df_guests["Judge B"] = array_judges[1]
        df_guests["Judge C"] = array_judges[2]
        df_judges = df_guests
        df_judges["Contestants"] = df_judges["Contestants"].apply(lambda x: x.replace(","," X "))
        df_judges["Season"] = i
        df_judges.drop(columns=["#"],inplace=True)
        df_judges.to_csv('csv/df_bbf.csv',mode="a",header=False,index=False, encoding="utf-8")
    except:
        pass
```

Com o arquivo das batalhas enriquecidos com todas as informações disponíveis, agora vamos carregá-lo para continuar com nossa análise.

```python
df_bbf = pd.read_csv("csv/df_bbf.csv")
df_bbf.head(5)
```

Agora que temos o dataset em nossas mãos podemos ver quantas vez Bobby Flay foi derrotado ou foi vencedor, para isso, pretendo criar uma nova coluna chamado "battle_status" que receberá dois valores "Knockout" para quando Bobby for o vencedor ou "Knocked Out" para quando ele for o perdedor (pensei em usar "beaten" e "beater", mas achei que ficaria muito parecido e pra ficar no tema de batalha resolvi usar knockout).

```python
df_bbf["battle_status"] = df_bbf["Winner"].apply(lambda x: "Knockout" if x == "Bobby Flay" else "Knocked Out")
df_bbf.head(5)
```

Agora podemos analisar quantas vitórias o Bobby tem no total

```python
df_bbf.groupby(["battle_status"]).size().sort_values(ascending=False)
```

Aqui nós podemos ver qual a dupla que mais visitou e venceu, nesse caso Giada de Laurentiis e Michael Symon são os maiores vencedores, com três vitórias. Para chegar nesses valores, apliquei um filtro na coluna battle_status para buscar apenas os valores "Koncked Out", agrupando pela coluna Guest(s), finalmente calculo a quantidade de ocorrência e ordeno do maior para o menor número de ocorrência, passando o parâmetro ascending=Flase na função sort_values. Feito isso, transformei o resultado em um dataframe e exibindo os 10 primeiros resultados, temos o retorno de quais são as duplas mais vencedoras.

```python
knocked = df_bbf.query("battle_status == 'Knocked Out'",engine="python").groupby(["Guest(s)"]).size().sort_values(ascending=False)
df_knocked = pd.DataFrame(knocked)
df_knocked = df_knocked.rename_axis("Guest(s)").reset_index()
df_knocked = df_knocked.rename(columns={0:"Victories"})
df_knocked.head(10)
# df_knocked_to_plt = df_knocked.query("Victories >= 2",engine="python")
# df_knocked_to_plt.plot.barh(x="Guest(s)",y="Victories")
```

Já para analisar os candidatos individualmente, criei um novo dataframe utilizando a função melt() para combinar as colunas Guest A e Guest B, tendo uma coluna só e apartir disso, aplicando os filtros correspondentes para chegar ao resultado do top 10 vencedores. E em primeiro lugar, temos a chef Alex Guarnaschelli, com um total impressionante de 21 vitórias. 

```python
guests_battles = df_bbf[["Guest A","Guest B","battle_status"]]
guests_battles = guests_battles.melt(id_vars=["battle_status"],value_vars=["Guest A","Guest B"],var_name=["Guest"])
guests_battles = guests_battles.query("battle_status == 'Knocked Out'",engine="python").groupby(["value"]).size().sort_values(ascending=False)
df_guests_battles = pd.DataFrame(guests_battles)
df_guests_battles = df_guests_battles.rename_axis("Guest").reset_index().rename(columns={0:"Victories"})
df_guests_battles.query("Victories >= 6",engine="python")
```
